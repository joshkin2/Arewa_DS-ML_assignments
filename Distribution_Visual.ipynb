```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("mushrooms.csv")
df.head()
```

```python
df["class"].hist()
plt.title("Mushroom Class Distribution")
plt.xlabel("Class")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.countplot(x="cap-color", data=df)
plt.title("Mushroom Cap Color Distribution")
plt.xlabel("Class")
plt.ylabel("Frequency")
plt.show()
```
