```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load dataset
path= r"birds.csv"
df= pd.read_csv(path)
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df["Name"]= df["Name"].str.title()
df.head()

```

```python
ross_goose= df[df["Name"]=="Ross'S Goose"]
ross_goose.head()
```

```python
plt.figure(figsize=(8,5))
plt.plot(["Min Body Mass","Max Body Mass"],[ross_goose["MinBodyMass"].values[0],ross_goose["MaxBodyMass"].values[0]],marker='o',linestyle="-",color="r")

plt.title("Ross's Goose Body Mass")
plt.xlabel("Body Mass(g)")
plt.ylabel("Weight")
plt.grid(True)
plt.show()
```

```python
geese= df[df["Genus"]== "Dendrocygna"]
plt.figure(figsize=(10,6))
sns.scatterplot(data=geese,x="MaxLength",y="MaxWingspan",hue="Name",s=100)

plt.title("Dendrocygna")
plt.xlabel("Max Length")
plt.ylabel("Max Wingspan")
plt.grid(True)
plt.legend(bbox_to_anchor=(1.05,1),loc="upper left")
plt.show()
```

```python
goose_names= geese["Name"]
min_mass= geese["MinBodyMass"]
max_mass= geese["MaxBodyMass"]
x= range(len(goose_names))

plt.figure(figsize=(10,6))
plt.bar(x,min_mass,label="Min Body Mass",color="r")
plt.bar(x,max_mass- min_mass,bottom=min_mass,label="Max Body Mass",color='lightblue')

plt.xticks(x,goose_names,rotation=45)
plt.ylabel("Body Mass (g)")
plt.title("Dendrocygna Body Mass")
plt.legend()
plt.tight_layout()
plt.show()
```
